```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

If on an average, 6 customers arrive every 2 minutes at a bank during the busy hours of working;<br/> (a) what is the probability that exactly 4 customers arrive in a given minute?<br/>(b) what is the probability that more than 3 customers will arrive in a given minute?

```python
# lambda : parameter of Poisson
avg_rate = 3 # 6 customers every 2 mins => 3 customers per min
```

```python
# Bounding array as no upper-bound given
n = np.arange(0, 20)
```

```python
poisson = stats.poisson.pmf(n, avg_rate)
poisson
```

```python
# Probability of exactly 4 customers arriving in a given min
poisson[4]
```

Approx. 17% chance is there that 4 customers arrive in a given min.

```python
# Probability that >3 customers arrive in a given min
1-(poisson[0] + poisson[1] + poisson[2] + poisson[3])
```

About 35% chance is that more than 3 customers arrive in a min.

```python
plt.plot(n, poisson, 'o-')
plt.title("Poisson: $\lambda$ = %i" % avg_rate)
plt.xlabel("No. of customers arriving in a min")
plt.ylabel("Probability of number of customers arriving in a min")
plt.show()
```
