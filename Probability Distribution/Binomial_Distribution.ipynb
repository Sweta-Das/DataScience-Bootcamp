```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

A bank issues credit cards to customers under the scheme of Master Card. Based on the past data, the bank has found out that 60% of all accounts pay on time following the bill. If a sample of 7 accounts is selected at random from the current database, construct the binomial distribution of accounts paying on time.

```python
# No. of trials (samples)
n = 7

# Proportion of customers who pay the bill on time
p = 0.6
```

```python
# No. of successes in each trial (Sample space)
k = np.arange(0, 8)
k
```

```python
# Probability of mass function
binomial = stats.binom.pmf(k, n, p)
binomial
```

It shows that:<Br/>
No account will pay its bill on time = 0.163 percent<Br/>
1 acc. will pay its bill on time = 1.702 percent <br/>
2 accounts will pay its bill on time = 7.741 percent <br/>
and so on.

```python
# Visualizing the distribution
plt.plot(k, binomial, 'o-')
plt.title("Binomial: n = %i, p = %.2f," % (n, p), fontsize = 15)
plt.xlabel("No. of successes")
plt.ylabel("Probability of successes")
plt.show()
```
